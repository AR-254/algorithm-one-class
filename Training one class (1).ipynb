```python
#import libraries
import pandas as pd
import numpy as np
import joblib
import os
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
import time

# Base path
base_path = r"C:\Users\ariam\Downloads\splits"

folders = ["Small", "Average", "Large"]

# columns to remove to prevent data leakage and avoid introducing bias
drop_columns = [
    'withdrew',
    'exam_score'
]

for folder in folders:
    path = os.path.join(base_path, folder)

    print(f"\n Processing {folder} dataset...")

    # Load data
    df = pd.read_csv(os.path.join(path, "X_train.csv"))

    # Drop leakage columns if exist
    df = df.drop(columns=[col for col in drop_columns if col in df.columns])

    # Ensure all values are numeric
    df = df.astype(float)

    print("Dataset shape:", df.shape)

    # Scaling 
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df)

    # One-Class SVM Model 
    model = OneClassSVM(
        kernel='rbf',
        gamma='auto',
        nu=0.05
    )

    #  Start training time
    start_time = time.perf_counter()

    model.fit(X_scaled)

    #  End training time
    end_time = time.perf_counter()
    
    print(f" Training time for {folder}: {training_time:.6f} seconds")

    training_time = end_time - start_time

    print(f" Model trained for {folder}!")

    # Save model & scaler 
    model_path = os.path.join(base_path, f"{folder}_oneclass_svm_model.pkl")
    scaler_path = os.path.join(base_path, f"{folder}_scaler.pkl")

    joblib.dump(model, model_path)
    joblib.dump(scaler, scaler_path)

    print(f" Model saved at: {model_path}")
    print(f" Scaler saved at: {scaler_path}")

print("\n All datasets processed successfully!")
```
